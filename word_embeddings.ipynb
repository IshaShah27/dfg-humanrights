```python
from gensim.models import Word2Vec
  
import nltk
import numpy as np 
  
from sklearn import cluster
from sklearn import metrics
import sklearn
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.metrics.pairwise import euclidean_distances
```

```python
from tqdm import tqdm
from tqdm.notebook import tqdm_notebook

tqdm_notebook.pandas()
```

```python
def preprocess(df):
    df.dropna(subset=['data_words'], inplace=True)
    df['data_words']=df['data_words'].progress_apply(lambda x: list(x.split(" ")) )
    data_words=df['data_words'].tolist()
    return data_words
```

```python
data_words=preprocess(documents)
```

```python
def paragraph_vectorizer(sent, model):
    sent_vec =[]
    numw = 0
    for w in sent:
        try:
            if numw == 0:
                sent_vec = model[vocab_to_int[w]]
            else:
                sent_vec = np.add(sent_vec, model[vocab_to_int[w]])
            numw+=1
        except:
            pass
    if len(sent_vec) == 300:

        return np.asarray(sent_vec) / numw
    else:
        return None
  
  
```

```python
X=[]
for idx, paragraph in tqdm(enumerate(data_words)):
    result=paragraph_vectorizer(paragraph, embed)
    if result is None:
        print('No Vector for Paragraph: {a}'.format(a=idx))
    else:
        X.append(result)
X=np.stack(X, axis=0 )
```

```python
import time
```

```python
t = time.process_time()
#kmeans = MiniBatchKMeans(n_clusters=10, random_state=0,verbose=5,compute_labels=True, batch_size=10000).fit(X)
kmeans = KMeans(n_clusters=10, random_state=0,verbose=5).fit(X)

center = kmeans.cluster_centers_
distances = euclidean_distances(X,center)
elapsed_time = time.process_time() - t
print('Elapsed_time: {a}'.format(a=elapsed_time))
```
